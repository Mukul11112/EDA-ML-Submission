## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Check missing values
df.isnull().sum().sort_values(ascending=False)

# Drop columns with high missing values
df.drop([
    'connected_handling_time',
    'Customer_City',
    'Product_category',
    'Item_price',
    'Customer Remarks'
], axis=1, inplace=True)

# Convert date columns
df['Issue_reported at'] = pd.to_datetime(
    df['Issue_reported at'],
    dayfirst=True,
    errors='coerce'
)

df['issue_responded'] = pd.to_datetime(
    df['issue_responded'],
    dayfirst=True,
    errors='coerce'
)

df['Survey_response_Date'] = pd.to_datetime(
    df['Survey_response_Date'],
    dayfirst=True,
    errors='coerce'
)

# Create Resolution Time Feature
df['Resolution_Time_Hours'] = (
    df['issue_responded']
    - df['Issue_reported at']
).dt.total_seconds()/3600

# Remove negative resolution time records
df = df[df['Resolution_Time_Hours'] >= 0]

# Check duplicates
df.duplicated().sum()
```

### What all manipulations have you done and insights you found?


### Data Manipulations Performed

1. Checked the dataset for missing values and identified columns with a large number of null values.
2. Removed columns such as **Customer City, Product Category, Item Price, Connected Handling Time, and Customer Remarks** because they contained a very high percentage of missing values.
3. Converted date-related columns (**Issue_reported at, issue_responded, Survey_response_Date**) into datetime format for time-based analysis.
4. Created a new feature called **Resolution_Time_Hours** by calculating the difference between issue reporting time and issue response time.
5. Identified and removed records with negative resolution times, as such values are not practically possible and indicate data inconsistencies.
6. Checked for duplicate records and found that the dataset contained no duplicate entries.

### Key Insights Found

1. The dataset initially contained significant missing values in several columns.
2. No duplicate records were present in the dataset.
3. Customer satisfaction scores (CSAT Score) are available for all records and can be used as the target variable.
4. Resolution time was successfully derived and can be used to analyze support efficiency.
5. After cleaning and feature engineering, the dataset became more suitable for Exploratory Data Analysis (EDA) and Machine Learning modeling.
6. The cleaned dataset can help identify factors that influence customer satisfaction and support performance.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): Resolution Time has no significant relationship with CSAT Score.

Alternative Hypothesis (H₁): Resolution Time has a significant relationship with CSAT Score.

#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(
    df['Resolution_Time_Hours'],
    df['CSAT Score']
)

print("Correlation:", corr)
print("P-value:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test

##### Why did you choose the specific statistical test?

Pearson Correlation is suitable because both Resolution Time and CSAT Score are numerical variables. It helps determine whether a significant linear relationship exists between them.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀ (Null Hypothesis): Average CSAT Scores are the same across all Agent Shifts.

H₁ (Alternative Hypothesis): Average CSAT Scores differ across Agent Shifts.

#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import f_oneway

morning = df[df['Agent Shift']=='Morning']['CSAT Score']
evening = df[df['Agent Shift']=='Evening']['CSAT Score']
night = df[df['Agent Shift']=='Night']['CSAT Score']

f_stat, p_value = f_oneway(morning, evening, night)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA Test

##### Why did you choose the specific statistical test?

One-Way ANOVA was chosen because it compares the mean CSAT Scores across multiple groups (Morning, Evening, and Night shifts). It helps determine whether agent shift has a significant impact on customer satisfaction.

H₁ (Alternative Hypothesis): Issue Category has a significant effect on CSAT Score.


#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import f_oneway

groups = [group['CSAT Score'].values
          for _, group in df.groupby('category')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA Test

##### Why did you choose the specific statistical test?

One-Way ANOVA was chosen because the dataset contains multiple issue categories, and the objective is to compare the mean CSAT Scores across these groups. This test helps determine whether customer satisfaction differs significantly among issue categories.
